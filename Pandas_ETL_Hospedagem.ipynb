```python
import pandas as pd
```

```python
dados = pd.read_json('/content/dados_hospedagem.json')
```

```python
dados.head()
```

```python
dados = pd.json_normalize(dados['info_moveis'])
```

```python
dados.head()
```

por esse arquivo, vamos entender o que cada coluna traz de informação.

    avaliacao_geral: refere-se à média de notas dadas para a avaliação da hospedagem no imóvel.
    experiencia_local: descreve as experiências oferecidas durante a hospedagem no imóvel.
    max_hospedes: informa a quantidade máxima de hóspedes que o local permite.
    descricao_local: descreve o imóvel.
    descricao_vizinhanca: descreve a vizinhança ao redor do imóvel.
    quantidade_banheiros: informa a quantidade de banheiros disponíveis.
    quantidade_quartos: informa a quantidade de quartos disponíveis.
    quantidade_camas: informa a quantidade de camas disponíveis.
    modelo_cama: informa o modelo de cama oferecido.
    comodidades: informa as comodidades oferecidas pelo imóvel.
    taxa_deposito: informa a taxa de depósito mínima para segurança de hospedagem.
    taxa_limpeza: informa a taxa cobrada para o serviço de limpeza.
    preco: refere-se ao preço base a ser cobrado pela diária no imóvel.

```python
colunas = list(dados.columns)
colunas
```

```python
dados = dados.explode(colunas[3:])
```

```python
dados
```

```python
dados.reset_index(inplace=True, drop=True)
dados
```

```python
dados.info()
```

```python
#Converter colunas numericas
import numpy as np

```

```python
dados['max_hospedes'] = dados['max_hospedes'].astype(np.int64)
```

```python
dados.info()
```

```python
colunas_numericas = [ 'quantidade_banheiros', 'quantidade_quartos', 'quantidade_camas']
dados[colunas_numericas] = dados[colunas_numericas].astype(np.int64)
```

```python
dados.info()
```

```python
dados['avaliacao_geral'] = dados['avaliacao_geral'].astype(np.float64)
```

```python
dados.head()
```

```python
colunas_numericas = [ 'taxa_deposito', 'taxa_limpeza']
dados[colunas_numericas] = dados[colunas_numericas].applymap(lambda x: x.replace('$','').replace('.','').replace(',', '.').strip())
dados[colunas_numericas] = dados[colunas_numericas].astype(np.float64)
```

```python
dados.info()
```

```python
dados.head()
```

```python
dados['preco'] = dados['preco'].apply(lambda x: x.replace('$','').replace('.','').replace(',', '.').strip()).astype(np.float64)
```

```python
dados.head()
```

Tokenização das descrições

```python
dados['descricao_local'] = dados['descricao_local'].str.lower()
dados['descricao_local']
```

```python
# Retirar simbolos não relevantes
dados['descricao_vizinhanca']=dados['descricao_vizinhanca'].str.replace('[^a-zA-Z0-9\-\']',' ',regex=True)
```

```python
# Retirar hifens soltos (menos os de palavras compostas)
dados['descricao_vizinhanca']=dados['descricao_vizinhanca'].str.replace('(?<!\w)-(?!\w)',' ',regex=True)
dados['descricao_vizinhanca']
```

```python
dados['descricao_local'] = dados['descricao_local'].str.split()
dados['descricao_local']
```

```python
dados['comodidades'] = dados['comodidades'].str.replace('\{|}|\"','',regex=True)
dados['comodidades']
```

```python
dados['comodidades'] = dados['comodidades'].str.split(',')
dados['comodidades']
```

```python
dados['descricao_vizinhanca']=dados['descricao_vizinhanca'].str.lower()
dados['descricao_vizinhanca']=dados['descricao_vizinhanca'].str.replace('[^a-zA-Z0-9\-\']',' ',regex=True)
dados['descricao_vizinhanca']=dados['descricao_vizinhanca'].str.replace('(?<!\w)-(?!\w)',' ',regex=True)
dados['descricao_vizinhanca']=dados['descricao_vizinhanca'].str.split()
```

```python
dados['descricao_vizinhanca']=dados['descricao_vizinhanca'].apply(lambda x: [palavra for palavra in x if len(palavra)>2])
dados['descricao_vizinhanca']
```

```python
#Remover palavras irrelevantes
palavras_irrelevantes = ['one','two','but','the','this','their']
dados['descricao_vizinhanca']=dados['descricao_vizinhanca'].apply(lambda x: [palavra for palavra in x if (palavra not in palavras_irrelevantes)])
dados['descricao_vizinhanca']
```
